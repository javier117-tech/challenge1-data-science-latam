### Importación de datos



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#suma de ingresos de cada tienda
suma1 = tienda["Precio"].sum()
suma2 = tienda2["Precio"].sum()
suma3 = tienda3["Precio"].sum()
suma4 = tienda4["Precio"].sum()
#Etiquetas y valores
tiendas = ["Tienda 1","Tienda 2","Tienda 3","Tienda 4"]
ingresos = [suma1,suma2,suma3,suma4]
```

```python
print('El total de ingresos de la tienda_1 fue: ',suma1)
print('El total de ingresos de la tienda_2 fue: ',suma2)
print('El total de ingresos de la tienda_3 fue: ',suma3)
print('El total de ingresos de la tienda_4 fue: ',suma4)
```

```python
#Creacion de grafico de barras
plt.bar(tiendas, ingresos, color=['blue','orange','green','red'])
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos Totales')
plt.show()
```

La tienda 1 tiene el ingreso más alto con 1150880400 y la tienda 4 el ingreso más bajo 1038375700

# 2. Ventas por categoría

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # Crear una cuadrícula de 2x2 para los gráficos
cantidad = tienda["Categoría del Producto"].value_counts()
print(cantidad)
cantidad.plot.pie(ax=axes[0, 0], autopct='%1.1f%%', title="Tienda 1")
print()
cantidad2 = tienda2["Categoría del Producto"].value_counts()
print(cantidad2)
cantidad2.plot.pie(ax=axes[0, 1], autopct='%1.1f%%', title="Tienda 2")
print()
cantidad3 = tienda3["Categoría del Producto"].value_counts()
print(cantidad3)
cantidad3.plot.pie(ax=axes[1, 0], autopct='%1.1f%%', title="Tienda 3")
print()
cantidad4 = tienda4["Categoría del Producto"].value_counts()
print(cantidad4)
cantidad4.plot.pie(ax=axes[1, 1], autopct='%1.1f%%', title="Tienda 4")
plt.tight_layout()  # Ajustar el diseño para evitar que los gráficos se superpongan
plt.show()
```

Respecto a las categorías de productos de las 4 tiendas, las 3 primeras vendieron la misma cantidad de 2359 productos en diferentes proporciones y la cuarta 2358 con distintas proporciones. En la tienda 1 la categoría más vendido fueron los muebles con el 19.7% y el menos los artículos para el hogar con el 7.2%, en la tienda 2 la categoría más vendida fueron los muebles con el 18.7%, el menos vendido fueron los artículos para el hogar con el 7.7%, en la tienda 3 la categoría más vendida fueron los muebles con un 21.2% y el menos vendido los artículos para el hogar 7.5% y los instrumentos musicales 7.5%, en la tienda 4 los muebles fueron los más vendidos con un 20.4% y los menos vendidos los instrumentos musicales 7.2%.


# 3. Calificación promedio de la tienda


```python
calificacion_promedio =tienda['Calificación'].mean()

calificacion_promedio_2 =tienda2['Calificación'].mean()

calificacion_promedio_3 =tienda3['Calificación'].mean()

calificacion_promedio_4 =tienda4['Calificación'].mean()
```

```python
tiendas = ["Tienda 1","Tienda 2","Tienda 3","Tienda 4"]
calificacion_promedio_tienda = [calificacion_promedio,calificacion_promedio_2,calificacion_promedio_3,calificacion_promedio_4]
```

```python
print('La calificación promedio de la tienda uno es : ',round(calificacion_promedio,2))
print('La calificación promedio de la tienda uno es : ',round(calificacion_promedio_2,2))
print('La calificación promedio de la tienda uno es : ',round(calificacion_promedio_3,2))
print('La calificación promedio de la tienda uno es : ',round(calificacion_promedio_4,2))
```

```python
plt.plot(tiendas, calificacion_promedio_tienda, marker='o', color='green', linestyle='--')
plt.title("Calificacion Promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificacion promedio")
plt.grid()
plt.show()
```

En la gráfica de puntos la calificación más baja fue la de la tienda 1 con un 3.98 y la más alta fue la tienda 3 con un 4.05.

# 4. Productos más y menos vendidos

```python
# Contar la frecuencia de cada producto
conteo_productos = tienda['Producto'].value_counts()
conteo_productos2 = tienda2['Producto'].value_counts()
conteo_productos3 = tienda3['Producto'].value_counts()
conteo_productos4 = tienda4['Producto'].value_counts()

```

```python
max_ventas = conteo_productos.max()
productos_mas_vendidos = conteo_productos[conteo_productos == max_ventas].index.tolist()

min_ventas = conteo_productos.min()
productos_menos_vendidos = conteo_productos[conteo_productos == min_ventas].index.tolist()

max_ventas2 = conteo_productos2.max()
productos_mas_vendidos2 = conteo_productos2[conteo_productos2 == max_ventas2].index.tolist()

min_ventas2 = conteo_productos2.min()
productos_menos_vendidos2 = conteo_productos2[conteo_productos2 == min_ventas2].index.tolist()

max_ventas3 = conteo_productos3.max()
productos_mas_vendidos3 = conteo_productos3[conteo_productos3 == max_ventas3].index.tolist()

min_ventas3 = conteo_productos3.min()
productos_menos_vendidos3 = conteo_productos3[conteo_productos3 == min_ventas3].index.tolist()

max_ventas4 = conteo_productos4.max()
productos_mas_vendidos4 = conteo_productos4[conteo_productos4 == max_ventas4].index.tolist()

min_ventas4 = conteo_productos4.min()
productos_menos_vendidos4 = conteo_productos4[conteo_productos4 == min_ventas4].index.tolist()
```

```python
resultados = {
    "productos_mas_vendidos": {producto: max_ventas for producto in productos_mas_vendidos},
    "productos_menos_vendidos": {producto: min_ventas for producto in productos_menos_vendidos}
}
resultados2 = {
    "productos_mas_vendidos": {producto: max_ventas2 for producto in productos_mas_vendidos2},
    "productos_menos_vendidos": {producto: min_ventas2 for producto in productos_menos_vendidos2}
}
resultados3 = {
    "productos_mas_vendidos": {producto: max_ventas3 for producto in productos_mas_vendidos3},
    "productos_menos_vendidos": {producto: min_ventas3 for producto in productos_menos_vendidos3}
}
resultados4 = {
    "productos_mas_vendidos": {producto: max_ventas4 for producto in productos_mas_vendidos4},
    "productos_menos_vendidos": {producto: min_ventas4 for producto in productos_menos_vendidos4}
}
```

```python
# Imprimir los resultados
print("Productos más vendidos:", productos_mas_vendidos, "con", max_ventas, "ventas cada uno.")

print("Productos menos vendidos:", productos_menos_vendidos, "con", min_ventas, "ventas cada uno.")
```
